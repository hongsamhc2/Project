```
import matplotlib.font_manager as fm
path = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf'  # 설치된 나눔글꼴중 원하는 녀석의 전체 경로를 가져오자
font_name = fm.FontProperties(fname=path, size=10).get_name()
print(font_name)
plt.rc('font', family=font_name)
fm._rebuild()
plt.figure()
df_result['소계'].plot(kind='barh',grid=True,figsize=(10,10))
```

```
df_result.sort_values(by='전체세대',inplace=True)
```

```
plt.figure(figsize=(20,10))
plt.scatter(df_result['전체세대'],df_result['소계'],s=50)
plt.xlabel('전체세대수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```
df_result.info()
```

```
df_result['전체세대'] = df_result['전체세대'].apply(lambda x : x.replace(',',''))
```

```
df_result['전체세대'] = df_result['전체세대'].astype(float)
```

```
df_result_max = df_result['전체세대'].max()
df_result_min = df_result['전체세대'].min()
```

```
cp1 = np.polyfit(df_result['전체세대'],df_result['소계'],1)
f1 = np.poly1d(cp1)
fx = np.linspace(df_result_min,df_result_max,100)
```

```
plt.figure(figsize=(20,10))
plt.scatter(df_result['전체세대'],df_result['소계'],s=50)
plt.plot(fx,f1(fx),ls='dashed',lw=3,color='g')
plt.xlabel('전체세대수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```
df_result['CCTV-세대 차'] = abs(df_result['소계'] - f1(df_result['전체세대']))
```

```
df_result['CCTV-세대 차']
```

```
df_diff = df_result.sort_values(by='CCTV-세대 차',inplace=False)
```

```
plt.figure(figsize=(14,10))
plt.scatter(df_result['전체세대'],df_result['소계'],
            s = 50,c = df_result['CCTV-세대 차'])
plt.plot(fx,f1(fx),ls='dashed',lw=3,color='g')

for _ in range(10):
  plt.text(df_diff['전체세대'][_]*1.02,df_diff['소계'][_]*0.98,
           df_diff.index[_],fontsize=15)
plt.xlabel('전체세대')
plt.ylabel('세대당 비율')
plt.colorbar()
plt.grid()
plt.show()
```
